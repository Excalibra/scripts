# Reading /nDong Degang: On the Reasons for the Decline in the Prestige of Marxism

## Negative Factors

**1. The Blow from the Failure of the Soviet-Style Socialist System**

Although unfamiliar with and having not studied Soviet history, this section of the article contains six key phrases: "KGB Group," "Military-Industrial Complex," "Dnepr Clique," "Crown Princes," "Secretaries' Party," "Second-Generation Officials," as well as “A small group of ruling party leaders monopolizing public power, lacking democracy, rule of law, and freedom. The ruling party imposes strict ideological control, adheres to leaders’ instructions as the norm, demands uniform thinking among citizens, allowing ‘one flower’ to bloom while ‘a hundred flowers’ wither, stifling innovation in thought and culture.” This is quite thought-provoking.

The failure of Marxism and the Soviet-style socialist system is a complex issue involving a combination of factors. Although Marxism pursues the ideal of benefiting the vast majority without special privileges, in practice, issues like power centralization, corruption, and economic planning failures emerged. The failure of the Soviet-style socialist system dealt a major blow to the prestige of Marxism.

**2. The Hypocrisy of Some Communist Officials and Marxist Theorists Severely Damaged Marxism's Reputation**

In *My Confession*, Yeltsin recalled the daily life of the Soviet privileged class and made this observation: “The higher you climb on the career ladder, the more luxurious the treatment you enjoy... If you succeed in reaching the pinnacle of the party's power pyramid, you get everything—you enter a communist state... Communism can indeed exist for a privileged few.” [<span id="fn1">(1996, Oriental Publishing House)</span>](#fn1-black).

Despite high-profile promotion of communism, collectivism, science, and morality, their behavior exhibited extreme individualism, authoritarianism, and decadent hedonism. Scandals exposed by the media, such as mistresses turning against them, illegitimate children, and "hush money" payments, made a mockery of their high-flown rhetoric about Marxism, reducing it to a political tool for embellishing reality, flattering superiors, deceiving the public, and self-delusion.

**3. Many Do Not Understand, Use, or Apply Marxism to Solve Practical Problems**

The *Collected Works of Marx and Engels* contains tens of millions of words, and due to specific historical contexts, national conditions, and linguistic barriers in translation, these works are not easy to comprehend. For ordinary workers and peasants preoccupied with making a living, delving deeply into these texts is even more challenging. Even among leading cadres and those who have read some works, truly grasping the essence and spirit of Marxism is a significant challenge.

> How can Marxism's prestige not be affected when people don’t understand it, can’t use it, and fail to solve real problems with it, yet still often invoke its name?

**4. The Denial of Marxist Doctrine by Some People Has Also Played a Role**

This is straightforward and requires no elaboration (omitted).

## Positive Factors

**1. Horizontally, Reform and Opening-Up Broadened Our Theoretical Horizons, Reducing Marxism's Relative Importance**

Horizontal: Timeline, China’s development relative to other countries’ development.

Excerpt from the first two paragraphs:

> In the past, we absolutized and sanctified Marxism, believing that Marx and Engels’ words were “absolute truth,” universally applicable in all contexts. Marxism became synonymous with science, truth, justice, nobility, and all things good, the pinnacle of human thought and culture. This exclusivity led to behaviors and attitudes that rejected any questioning of Marxist doctrines or its contemporary representatives—the leaders’ speeches—and excluded the valid contributions of other ideologies.
>
> With the deepening of reform and opening-up, people’s horizons have expanded, gaining a world-historical perspective. They have observed that most countries worldwide are not guided by Marxism but have still developed; some have even achieved rapid progress, creating extraordinary human achievements. Frequently, leaders ask: Many countries do not follow Marxism, yet they thrive and, in many cases, surpass us in development—are their ideologies entirely wrong? Obviously not, as their achievements suggest otherwise.

**2. Vertically, We Created Theories Such as Socialism with Chinese Characteristics, Reducing Marxism's Relative Importance**

Vertical: Self-created content and dimensions.

Over time, as society changes and develops, more ideological and academic achievements emerge, reducing the weight and relative standing of Marxist doctrine. This is an inevitable historical trend.

Excerpt from the concluding paragraph:

> Mao Zedong once said: “In a few thousand years, Marx will be regarded as we now regard Confucius.” This statement is insightful. Recall that in feudal China, many believed, “If Confucius had not been born, the world would have been in perpetual darkness.” History has shown the absurdity of this claim. While Confucius made great contributions, his teachings alone cannot address modern needs. Similarly, Marx cannot answer all questions future generations encounter; it is up to them to devise solutions. Practice, science, and the times evolve; successors will inevitably create new theories, surpassing Marxism. All scientific theories are stepping stones for understanding the world, inherently limited and subject to analysis, development, and eventual replacement.

**3. People Are Now Increasingly Able to Take a Scientific Approach to Marxism, Causing Its Prestige to Recede from Its Peak**

Excerpt from the last fifth, third, and final parts:

> Abroad, Marxist scholars increasingly analyze Marxism scientifically. Chen Xueming summarized: "Western Marxist theorists propose five views on the main characteristics of Marxism: First, Marxism is not monolithic but a pluralistic system..."

> Some argue that Marxism primarily critiques capitalism, meaning it remains relevant as long as capitalism exists. While this holds some truth, modern developed capitalism has evolved significantly, integrating elements of socialism, such as social welfare systems and macroeconomic regulation. These changes render Marx’s original conclusions outdated.

> Thus, Marxism has been “demystified,” transitioning from its inflated pedestal to being part of the diverse intellectual and cultural landscape of the contemporary world and China.

## Lessons Learned

(Full section excerpted)

> The main lessons to be drawn from the decline in Marxism’s prestige can be summarized as follows:
>
> Negative factors teach us: First, the Soviet-style socialist system must undergo thorough reforms to enhance democracy and the rule of law, particularly addressing issues of centralization, bureaucratic privileges, and corruption. Second, Communist leaders and theorists must internalize Marxism’s universal truths and other civilizational achievements to genuinely practice them. Third, efforts must focus on applying Marxism to solve real-world problems, advancing China's modernization and improving people’s livelihoods. Fourth, criticisms of Marxism should be met with reasoned arguments, not political labeling.
>
> Positive factors teach us: First, a scientific approach to Marxism is essential—emphasizing its universal truths while avoiding dogmatism. Second, Marxism should be viewed within a global historical context, recognizing it as one among many intellectual peaks, complementing other achievements. Third, Marxism must evolve with practice and scientific advancements, with new theories emerging to supplement, enrich, and surpass it, ensuring continual intellectual growth.
