## Reflections on 'The Actual State and Guiding Strategies of Youth Online Patriotism – Taking the 'Little Pink' Group as an Example'

### Little Pink

According to [<span id="fn1">Niufan, 2022.9</span>](#fn1-black), the 'Little Pink' group has relatively simple patriotic feelings and enthusiasm for political participation. They tend to achieve national identity through emotional experiences. Members of this group show great differences in terms of economy and culture, generally lacking strict work experience in society. They find it difficult to think about issues within complex real-world contexts and are more accustomed to identifying 'friends and enemies' through language symbols, forming their cognitive elements through labeling. However, Niufan also pointed out that emotionalized discussions and labeling oppositions are not unique to the 'Little Pink' group, as this trend exists globally.

Niufan started from micro-level aspects such as economic consumption capacity, interests, hobbies, knowledge background, and generational characteristics, reflecting the actual state of most 'Little Pink' members, and reached the following macro conclusions:

1. Based on the urban middle class, the group image is diversified, including members living and studying abroad, coming from different online communities, and having diverse personalities.
2. Focused on emotional expression, their content output is entertainment-oriented, involving personification, idolization, familial connections, and content like 'Year Hare Affair' linking international recognition with emotional ties.
3. Emphasizing self-display, their language expression is symbolized through bullet screens, short videos, deconstruction, self-mockery, and a sense of belonging and self-worth.
4. Supported by loose aggregation, the organizational form is fluid (topic discussions create a sense of collective identity, but participants often return to their individual lives afterward, exhibiting the typical characteristics of fluid crowds or "mob-ility").

However, I noticed a missing factor. The 'Little Pink' living abroad often observe and experience different social environments, which makes them more attentive to social phenomena that contradict their values and background. They notice societal issues more sharply, perhaps because these problems become more obvious in contrast to the foreign society they are in. Emotions tied to national honor and personal confidence influence their self-identity, binding them to their country emotionally and motivating them to defend national honor and values. Essentially, the 'Little Pink' overseas pay more attention to social issues in their environment and link national and personal honor to their self-identity, driving them to contribute to national progress through patriotism.

Niufan also explained this in tracing the causes of the 'Little Pink'. He then proposed strategies to guide the 'Little Pink' and strengthen patriotic propaganda:

1. Educators should closely follow the ideological trends and behavioral patterns of the 'Little Pink', using scientific stratification and classification to guide them, particularly focusing on ideological and political education for key figures in the group.
2. Sensitive words and emoticons should be selectively screened and filtered, and the activities of the 'Little Pink' should be monitored with careful analysis of public opinion.
3. Patriotic content should be actively planned from the youth perspective, increasing the immersive experience of propaganda content in daily life, subtly cultivating patriotic ideology.

Patriotism is a serious topic. The 'Little Pink' group expresses their patriotism through emotionally charged, personal experiences online, often lacking deep thought about history, the nation's future, and rational discourse. As Niufan points out, many 'Little Pink' members might be more in love with themselves, elevating personal ideals to expressions of national loyalty.

## Reflections on 'Study of the Realistic Considerations and Ways to Improve the Social Responsibility of the 'Little Pink''

### Conclusion and Analysis of the 'Little Pink' Group

According to [<span id="fn2">Liu Fang, 2019.4</span>](#fn2-black), 'On one hand, we need to enhance the 'Little Pink's political participation and new media literacy, fostering a sense of great nationhood to increase their capacity for bearing social responsibility. On the other hand, the media must play a role in correcting misunderstandings about the 'Little Pink' by promoting their social identity and transforming their attitude toward social responsibility. Lastly, government departments should step in to provide macro-level guidance, strengthen ideological education for the 'Little Pink', offer safe and reliable internet access, and further strengthen internet regulation, creating a clear online space for them.'

'Enhancing the 'Little Pink's political participation and new media literacy, fostering a sense of great nationhood' highlights their shortcomings in political knowledge and media literacy. The patriotism they express is more based on personal feelings. 'The media should correct misunderstandings and promote the social identity of the 'Little Pink'' reflects concerns about the emotional and singular expressions of patriotism.

'Government departments should provide macro-level guidance and strengthen ideological education for the 'Little Pink'', but overemphasis on this can lead to negative effects:

1. **Restricting Freedom of Speech**: Strengthening ideological education and internet regulation can excessively limit their freedom of expression, leading to a more controlled and uniform way of thinking.

2. **Suppressing Creativity and Innovation**: Excessive control may suppress diversified thinking and creativity, limiting exposure to different ideas and stifling innovation.

3. **Social Division and Conflict**: Overemphasizing control over the 'Little Pink' could deepen social divisions and conflicts, as other groups may feel marginalized.

4. **Infringing on Netizens' Autonomy**: While protecting teenagers from harmful information is necessary, overly strict internet regulation may infringe on users' autonomy and privacy.

Ideological education and internet regulation should listen to voices from all sides, fostering an open and inclusive environment that ensures individual rights and social stability are not compromised. The diversification and inclusive development of society encourage dialogue and understanding between different groups. Only by enhancing social identity and establishing trust mechanisms can phenomena of societal division and opposition be reduced. Solutions should include efforts to improve literacy and ideological education, ensure objective and impartial media reporting, achieve a balanced approach in government macro-regulation, and promote the diversified and inclusive development of society. In this way, a social environment that builds consensus and embraces diversity can be cultivated.

### Definition of the Age and Main Constituent Groups of 'Little Pink'

“The term “Little Pink” is a colloquial reference to a group of young, patriotic internet users. The “Little” signifies their youth, indicating that while they may be immature, they are full of energy and enthusiasm. The “Pink” refers to their use of a unique, playful online communication style, often employing cute language and a variety of adorable emojis. Additionally, their limited life experience contributes to a still-developing worldview. The “Red” symbolizes their strong ideological roots, as red in the Chinese context often represents the Communist Party of China. Here, “Red” reflects their fervent love for the Party, the country, and its leadership. [<span id="fn2">Liufang (April 2019)</span>](#fn2-black)

Liufang's definition of the 'Little Pink' group is accurate. She further subdivides this group based on different age ranges, social classes, and professional backgrounds. Within the 18-to-24 age bracket, students constitute a significant proportion of the 'Little Pink,' with most coming from working-class families. However, under the classification models popular before the Old Niufan framework, a considerable number of 'Little Pink' also hailed from urban middle-class families, which is worth noting. Regardless, these two groups share a common trait: the 'Little Pink' possess relatively strong social consumption power and purchasing ability while facing comparatively less social pressure. They are also beneficiaries and witnesses of China's growing national strength following the Reform and Opening-Up policy.

### The Causes and Self-Reflection of the “Little Pink” Phenomenon

Before delving into this, we must first understand what nationalism is, and then comprehend online nationalism, in order to clarify the causes and subsequent development of the 'Little Pink' phenomenon. A nation is often referred to as an 'imagined community.' The reason a nation is considered an imagined community is that, although people live in the same community, they do not personally know each other. Instead, they are connected through media, which creates shared experiences and acts as a bond to maintain emotional ties, forming a collective sense of emotion and experience. [<span id="fn3">Li Hongmei（2016.11）</span>](#fn3-black)。

The emergence of the 'Little Pink' group is closely related to online nationalism. It is shaped by a combination of factors such as the rise in China's comprehensive national strength and international influence, incidents like overseas anti-China movements, anti-Chinese sentiment, and the influence of these issues spreading and fermenting through the internet, all of which contribute to its birth, carrying strong political overtones.

There has been ongoing debate and controversy regarding the 'Little Pink,' along with the rapid development of self-media. The diversification of media content and platforms has led to an increase in the variety of information people browse, with a more vertical focus on content, while political news has shown signs of decline. At the same time, as the 'Little Pink' grow older and gain more social experience through work and life, they no longer hold the same views as in their youth. They begin to reconstruct their values and enter a stage of reflection.

### Post-'Little Pink'

The national identity of the 'Little Pink' has undergone a shift from being emotionally driven to being more rationally controlled. This is a typical characteristic of the 'post-Little Pink.'


Little A, born after 1995.<br>
Q: Do you admit to being a 'Little Pink'?<br>
A: Yes, but with a prefix: I am a rational, patriotic 'Little Pink.<br>
Q: Why add this qualifier?<br>
A: Because one of the biggest characteristics of the 'Little Pink' I know is being irrational, and I absolutely disagree with that. This is also a misunderstanding of the 'Little Pink' by society. I don't deny that there are irrational, 'brain-dead' 'Little Pink' in society, but we shouldn't generalize. At least, for me or the 'Little Pink' around me, we are still relatively rational. <br>
Q: Have you heard about the speech by the overseas students at the University of Maryland? <br>
A: Yes, actually when I first heard the news, I was quite angry, just like many netizens. But upon further thought, not everything she said was exaggerated or false. Objectively speaking, our country's air pollution situation is indeed quite serious, but it is also continuously improving, so there's no need to get too worked up over it. <br>
Q: As a member of the 'Little Pink,' how do you view your social responsibility?<br>
A: Overall, we are quite patriotic, both in our actions and our words. We are also proud of our country's development. However, in the process of having this patriotic sentiment, we still lack the ability to objectively and calmly analyze public opinion. This leads the public to question whether we can bear social responsibility. <br>

**The conversation with Little A touches on the identity of 'Little Pink' and their social responsibility.** Regarding the 'Little Pink' identity, the respondent acknowledges that they are a 'rational patriotic “Little Pink”' and points out that 'Little Pink' is misunderstood as being brain-dead, but they and the people around them are relatively rational. As for the speech by overseas students at the University of Maryland, the respondent initially felt angry but later realized that some points in the speech were not exaggerated or false, and also mentioned that the air pollution issue in our country is improving and should be faced calmly. When discussing the social responsibility of the 'Little Pink' group, the respondent believes that most people show patriotism but lack the ability to objectively and calmly analyze public opinion, which leads the public to question whether they can take on social responsibility.

Little B, born after 1990, male.<br>
Q: What do you think are the typical characteristics of 'Little Pink'? <br>
A: Personally, I think 'Little Pink' are relatively young, or immature? They also tend to have a lower level of education and lack judgment. It can generally be understood as fans who blindly worship the country. <br>
Q: Why make these judgments? Don't you mind clearing your name and getting rid of these unpleasant labels?<br> 
A: Take the Empire Bar incident for example. It was quite a sensation at the time, and social media platforms like Weibo, QQ, and Facebook played a big role. But what role did they really play? Was it just sending memes? Promoting socialist core values? It looks very childish and exaggerated when presented in a mocking way. Plus, many participants didn't even know why they were taking part, to be honest, they were just joining in for the fun. If that's the case, then how can others' opinions of them change?<br>
Q: So, what do you think 'Little Pink' can do to break out of this vicious cycle of identity?<br>
A: The main thing is to improve one's ability and quality. When dealing with political issues, they don't need to be so radical, but rather engage in dialectical analysis, as philosophy advocates.<br>

**The conversation with Little B touches on the characteristics and identity issues of the 'Little Pink' group.** The respondent believes that 'Little Pink' are generally young, with relatively low education levels, lacking judgment, and can be understood as blindly worshiping the country. The respondent cites the Empire Bar incident, noting that while it caused a sensation on social media, most participants did not understand the purpose behind it and were merely following the crowd for fun. The respondent believes such behavior seems childish and easily criticized. Regarding how to escape the negative labels of 'Little Pink,' the respondent suggests improving personal ability and quality, and approaching political issues with dialectical analysis, rather than being overly radical.

Little C, born after 1995, male, overseas student. <br>
Q: From what you know, are there many 'Little Pink' among the overseas students around you?<br>
A: Yes, because after going abroad, you feel that the foreign land is very unfamiliar, and then you start to want to return to your country.<br>
Q: Has your view of the motherland changed before and after studying abroad?<br>

In comparison, you notice the difference. For example, in your eyes, China is constantly developing and growing, but the foreign media descriptions are not always the same. It's not that they are fake news, but the perspectives and content are different.<br>
Q: So, in your opinion, how should 'Little Pink' express their patriotism appropriately?<br>
A: You should look at the world as a whole, not just focus on the domestic view. You need to engage more with the outside world and have the demeanor and vision that a big country should have.<br>

**The conversation with Little C explores the phenomenon of 'Little Pink' among overseas students, the change in their views on the motherland, and how to appropriately express patriotism.** The respondent mentions that the 'Little Pink' phenomenon is quite common among overseas students. The main reason is that in a foreign country, there is a feeling of homesickness and a desire to return to the homeland. Before and after studying abroad, the respondent's views on the motherland changed. Before going abroad, they received socialist education, and the news they encountered was mostly domestic. After studying abroad, they started to see China from the perspective of foreigners and noticed the difference between the two. This doesn't mean that foreign news is fake, but the perspectives and details are different. Regarding how to appropriately express patriotism, the respondent believes that one should look at the world and not just limit themselves to a domestic perspective, while engaging more with the outside world. At the same time, they should have the demeanor and vision that a great country should possess. **In summary, this dialogue revolves around the 'Little Pink' phenomenon among overseas students, including homesickness, the change in views on the motherland, and advice on how to appropriately express patriotism.**

### Guiding Little Pink to Serve the Country and Follow the Chinese Dream
'Little Pink' is a group with strong appeal and mobilization power. By making 'Little Pink' the reserve force for the socialist cause, they can actively support the Communist Party's leadership and promote mainstream values. At the same time, improving the internet regulation system can create a clean and clear online space for 'Little Pink,' allowing them to become leaders of ideological and political education online, evaluators of good and bad information, role models to foster patriotism, and major contributors to the construction of the national image, actively defending the image of the motherland. 

“Little Pink” applies what they have learned and know to social practice, serving the development and operation of the country and society, while supervising factors that are detrimental to social stability and national security. They fully utilize their initiative, influencing and inspiring more people, both online and offline, to engage in scientific and rational patriotic practices.

### A Brief Analysis of the Marxism College's Stance and Emotional Elements

I feel that the people in the Marxism College are quite sharp and have a broad perspective when analyzing sociological aspects. However, while they appear righteous and intent on creating a binary opposition of justice versus evil, in actual political environments, their attitudes and actions may be influenced by various factors, including the influence of the ruling party.

When the people in the Marxism College support the ruling party, they may maintain and support the party's policies and decisions. This could lead them to view the actions of the ruling party more positively in their analysis and defend it. They may emphasize the efforts and achievements of the ruling party, while offering less criticism of its negative effects and limitations. However, if the people in the Marxism College disagree with the ruling party's stance or hold opposing views, they may express differing opinions and offer criticism. In this case, they might highlight the problems and shortcomings of the ruling party's policies, suggest improvements, and support other political forces or social movements.

It should be made clear that political attitudes and behavior are influenced by multiple factors, not just by the college or ideological school to which one belongs. An individual's values, principles, experiences, and surrounding environment can all affect their political stance. Therefore, although people in the Marxism College may present themselves as embodiments of justice and righteousness, their attitudes and actions are still diverse and subject to change.

## Reflections on the Analysis of the 'Diba Expedition' Incident

At the end of 2015, Taiwanese artist Zhou Ziyu, while participating in a Korean variety show, waved the "Republic of China (ROC) flag". This incident was exposed by Taiwanese artist Huang An on January 8, 2016 via Sina Weibo, leading to a boycott by Chinese mainland netizens and advertisers. Subsequent nationalist events included Chinese netizens organizing themselves in virtual space under the name of 'Emperor Bar' (帝吧) to attack Facebook. They flooded the Facebook pages of Taiwan's Sanlih News, Apple Daily, and Liberty Times, and also hacked into the Facebook page of Taiwan's then 'presidential candidate' Tsai Ing-wen. The 'Emperor Bar' attackers used various images, texts, and memes to express their identity and achieve the goal of "cultural exchange".[<span id="fn3">Li Hongmei (November 2016)</span>](#fn3-black)。

### Taiwan Identity Topic

For a long time, the people of mainland China have hoped for more communication with the people of Taiwan, but geopolitical factors, different political systems, and the roles of the United States and Japan have caused Taiwan and mainland China's relationship to be full of political divergences. Taiwan's de-Sinicization is largely aimed at creating another kind of imagination, 'Taiwan for Taiwanese'. Through textbooks, media, and policies, they sever the historical ties with China in order to establish an independent political system. Moreover, despite Taiwan having a population that is 97% Han Chinese, the localization movement of recent decades has led to fewer people identifying as Chinese and more identifying as Taiwanese. For example, a public survey conducted by Taiwan's National Chengchi University in 2014 showed that 23.8% of Taiwanese people lean towards supporting Taiwan independence or hope for its immediate realization; 60.4% of the population only identify as Taiwanese; 32.7% identify as both Taiwanese and Chinese, but only 3.5% consider themselves as solely Chinese; the proportion identifying with Taiwan independence and Taiwanese identity reached its highest since 1994. [Li Hongmei (2016.11)]

Due to the 'Great Unity' education received from a young age, Chinese mainland citizens with a nationalist sentiment feel a strong sense of kinship with Taiwan. On the other hand, their frustration and resentment towards Taiwan's 'localization' and the growing distance from China's 'unification' policy causes cognitive dissonance.

### Summary of Li Hongmei's Conclusion

The article analyzes Chinese nationalism from a macro perspective, placing it in both international and domestic contexts. Nationalism plays a positive role in shaping the cohesion of the nation-state, and at the same time, on international issues, individuals turn to nationalism to vent their dissatisfaction with the current social situation, ease public grievances, and also protect government interests.

On the other hand, overseas scholars believe that nationalism is a tool used by elites to control the masses. Nationalism has the following characteristics: 'A sense of belonging is the only right, loyalty as the supreme obligation; dignity exists only in the collective glory; private interests and collective welfare are linked.' From this perspective, nationalism is essentially a tool of control and an ideology. We need to be cautious about when nationalism is used as a tool to suppress different voices and protect the interests of specific groups.

The internet, as a space for many citizens to express opinions, inevitably reflects certain issues and values of society as a whole. However, we must also recognize that the internet often amplifies extreme opinions. The voices online do not necessarily reflect public opinion. Drawing conclusions about Chinese nationalism purely from online discussions is incomplete and unobjective. Although online nationalism does indeed present some new characteristics, we must place the study of online nationalism within the broader social and cultural context. Online nationalism is largely a performance related to identity politics in the context of consumer culture and global geopolitics. This performance is not only rooted in our selective memory of history, but also an assertion of the position that Chinese people and China should have in the global context. It is a response to the setbacks China has encountered in the process of globalization (such as issues related to China's sovereignty, territorial integrity, and identity). It is a combination of domestic and international forces, political and economic cultures, and both official and grassroots factors. [<span id="fn3">Li Hongmei (2016.11)</span>](#fn3-black).
