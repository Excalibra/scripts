## Review of Online Lurking Research

### Concept Definitions and Research Progress

#### Definition of Digital Natives

The term "Digital Natives" was first introduced by Marc Prensky in 2001 to differentiate from "Digital Immigrants." Digital natives are those born, educated, and raised in a high-tech environment, immersed in the digital world from birth. In contrast, digital immigrants are older individuals who must learn digital technologies with difficulty or struggle. (Liu Jiang, 2013, p9)

#### Definition of Online Lurkers

According to Liu Jiang [<span id="fn1">(2013, p12-13)</span>](#fn1-black), online lurkers can be subdivided as follows:

**Passive Lurkers (P):** Traditional lurkers who strictly observe the "read-only" rule, accessing information without contributing or sharing content.

**Active Lurkers (A):** Individuals who refrain from creating or sharing content publicly but engage directly with creators via private channels, such as emails or instant messaging, to exchange opinions or acquire information.

**Occasional Lurkers (OL):** Users who are intermittently active, alternating between lurking and engaging due to personal or environmental factors. Their behavior can help community managers track user retention and loyalty.

**De-lurkers (DL):** Users who, after a period of lurking, start creating, sharing, or discussing content within a community due to specific triggers.

#### Progress in Lurking Research

Liu Jiang [<span id="fn1">(2013, p13-15)</span>](#fn1-black) provided a bibliographic sample and statistical summary: 32 relevant studies were identified, excluding one book chapter, resulting in 33 research papers. Lurking research is still in the exploratory stage.

### Motivation System for Lurking Research

Foreign researchers primarily employ surveys, interviews, and experiments to study external motivations. Extracted motivations are diverse. [<span id="fn1">(Liu Jiang, 2013, p18)</span>](#fn1-black)

#### Internal Motivations

Internal motivations explore factors influencing lurking behavior from within, such as psychology, personality traits, and knowledge skills. Internal motivations are highly autonomous, shaped by intrinsic psychological traits, emotions, and cognitive structures. Key aspects include:

1. Privacy and Security: Fear of privacy breaches or insecurity leads to lurking.
2. Personality Traits: Shyness and lack of confidence inhibit participation.
3. Time Constraints: Limited personal time results in "read-only" behavior.
4. Posting Commitment: Community does not meet personal expectations of belonging.
5. Knowledge and Skills: Lack of expertise or personal capability.

#### External Motivations

External motivations focus on external influences like human factors, organizational aspects, environmental (contextual) factors, and societal impacts. These motivations mainly involve economic, social, and technical aspects. Examples include:

1. Information Quality: Poor or overwhelming information reduces engagement.
2. Work Restrictions: Inappropriate group atmosphere or censorship.
3. Organizational Recognition: Lack of acceptance or unfriendly environment.
4. Technical Barriers: Early forum designs required technical skills.
5. Interaction Delays: Lack of prompt feedback or interaction.
6. External Rewards: Material or financial incentives, although limited by individual commitment.

#### Incentive Measures

Suggestions:

1. Strengthen awareness of security and advocate privacy protection.
2. Create discussion groups or "circles" for personalized interaction.
3. Encourage comprehensive topic discussions.
4. Develop professional zones and open "free discussion" areas.
5. Provide tutorial videos to build foundational knowledge.
6. Offer recognition, such as badges, personalized messages, or dedicated discussion groups.

Liu Jiang's Opinions:

1. Address diverse motivations with targeted incentives based on demographics.
2. Enhance external incentives and community interactions to reduce lurking.
3. Cultivate a sense of virtual community by fostering trust and harmony.
4. Conduct regular surveys to understand lurking motivations and dissatisfaction.


## Theoretical Model Construction and Hypothesis Proposal

### Theoretical Basis for Lurking Research

Early research focused on identifying legitimate lurkers and differentiating them from outsiders. With support from disciplines like sociology, social psychology, organizational behavior, and economics, lurking behaviors have been further classified. Liu Jiang [<span id="fn1">(2013, p20)</span>](#fn1-black) examined this evolution, citing studies like:

* Cliff Lampe: Explored the "social loafing" effect in forums.
* Rafaeli et al.: Applied social capital theory to analyze virtual community engagement.

### Theoretical Foundations for Model Construction

#### Theory of Planned Behavior

Proposed by Ajzen in 1991, this theory identifies five key elements: attitude, subjective norm, perceived behavioral control, behavioral intention, and actual behavior. (Liu Jiang, 2013, p27)

#### Social Exchange Theory

Formula: **Value = Rewards - Costs** (cooperation intention).

#### Social Capital Theory

Focuses on group expectations and shared trust within a community, exploring team coordination and social efficiency.

#### Social Identity Theory

Combines theories of planned behavior, social exchange, and social capital to understand virtual community dynamics.
